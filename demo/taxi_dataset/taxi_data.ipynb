# Load and fix data types

```python
import numpy as np
data = pd.read_csv("https://s3.amazonaws.com/nyc-tlc/trip+data/green_tripdata_2020-06.csv", parse_dates=True)
#data["lpep_pickup_datetime"] = pd.to_datetime(data["lpep_pickup_datetime"])
#data["lpep_dropoff_datetime"] = pd.to_datetime(data["lpep_dropoff_datetime"])
data["RatecodeID"] = data.RatecodeID.replace({
    1: "Standard Rate",
    2: "JFK",
    3: "Newark",
    4: "Nassau or Westchester",
    5: "Negotiated Fare",
    6: "Group Ride"
})
data["payment_type"] = data.payment_type.replace({
    1: "Credit Card",
    2: "Cash",
    3: "No Charge",
    4: "Dispute",
    5: "Unknown",
    6: "Voided Trip"
})
data["trip_type"] = data.trip_type.replace({
    1: "Street Hail",
    2: "Dispatch"
})
data = data.drop("ehail_fee", axis="columns")
data["id"] = np.arange(data.shape[0])
```
